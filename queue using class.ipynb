```python
class Queue:
    def __init__(self, max_size):
        self.queue = []
        self.max_size = max_size
        self.front = -1
        self.rear = -1

    def is_empty(self):
        return self.front == -1 and self.rear == -1

    def is_full(self):
        return self.rear == self.max_size - 1

    def add_element(self, data):
        if not self.is_full():
            if self.is_empty():
                self.front = self.rear = 0
            else:
                self.rear += 1
            self.queue.append(data)
            print(f"\n{data} is added to the queue\n")
        else:
            print("\nQueue is full\n\n")

    def delete_element(self):
        if not self.is_empty():
            print(f"\nDeleted element is {self.queue[self.front]}\n")
            self.front += 1
            if self.front > self.rear:
                self.front = self.rear = -1
        else:
            print("\nQueue is empty\n")

    def list_full(self):
        if self.is_full():
            print("\nQueue is full\n")
        else:
            print("\nQueue is not full\n")

    def list_empty(self):
        if self.is_empty():
            print("\nQueue is empty\n")
        else:
            print("\nQueue is not empty\n")

    def display(self):
        if not self.is_empty():
            print("\nQueue elements are: ")
            for i in range(self.front, self.rear + 1):
                print(self.queue[i], end=" ")
            print()
        else:
            print("\nQueue is empty\n")

size = eval(input("Enter the size of the queue: "))
my_queue = Queue(size)
print()
while True:
    print("1. Add element to the queue")
    print("2. Delete element from the queue")
    print("3. Check if the queue is full")
    print("4. Check if the queue is empty")
    print("5. Display the queue")
    print("6. Exit")
    choice = int(input("\nEnter your choice: "))

    if choice == 1:
        element = int(input("Enter the element to be added: "))
        my_queue.add_element(element)
    elif choice == 2:
        my_queue.delete_element()
    elif choice == 3:
        my_queue.is_full()
    elif choice == 4:
        my_queue.is_empty()
    elif choice == 5:
        my_queue.display()
    elif choice == 6:
        break
    else:
        print("\nInvalid choice. Please try again.\n")

print("\nThank you for using the Queue implementation.\n")
```
